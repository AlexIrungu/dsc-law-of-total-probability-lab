## Exercise 4

United Airlines operates flights from JFK to Amsterdam, to Brussels, and to Copenhagen. As you may know, flights are overbooked fairly often. Let's denote the probability of the flight to Amsterdam being overbooked equal to 40%, the probability of the flight to Brussels being overbooked equal to 25%, and the probability of the flight to Copenhagen being overbooked equal to 35%. You can assume that these events of overbooking are independent events.

- Compute the probability that all the flights are overbooked.
- Compute the probability of having at least one flight which is not overbooked.
- Compute the probability that exactly one flight is overbooked.

```python
# Your code here
# Probabilities of each flight being overbooked
P_Amsterdam_overbooked = 0.4
P_Brussels_overbooked = 0.25
P_Copenhagen_overbooked = 0.35

# Part 1: Probability all flights are overbooked
P_all_overbooked = P_Amsterdam_overbooked * P_Brussels_overbooked * P_Copenhagen_overbooked

# Part 2: Probability of at least one flight not being overbooked
P_Amsterdam_not_overbooked = 1 - P_Amsterdam_overbooked
P_Brussels_not_overbooked = 1 - P_Brussels_overbooked
P_Copenhagen_not_overbooked = 1 - P_Copenhagen_overbooked
P_all_not_overbooked = P_Amsterdam_not_overbooked * P_Brussels_not_overbooked * P_Copenhagen_not_overbooked
P_at_least_one_not_overbooked = 1 - P_all_not_overbooked

# Part 3: Probability that exactly one flight is overbooked
P_only_Amsterdam_overbooked = P_Amsterdam_overbooked * P_Brussels_not_overbooked * P_Copenhagen_not_overbooked
P_only_Brussels_overbooked = P_Brussels_overbooked * P_Amsterdam_not_overbooked * P_Copenhagen_not_overbooked
P_only_Copenhagen_overbooked = P_Copenhagen_overbooked * P_Amsterdam_not_overbooked * P_Brussels_not_overbooked
P_exactly_one_overbooked = P_only_Amsterdam_overbooked + P_only_Brussels_overbooked + P_only_Copenhagen_overbooked

# Print the results
print(f"Probability all flights are overbooked: {P_all_overbooked}")
print(f"Probability of at least one flight not being overbooked: {P_at_least_one_not_overbooked}")
print(f"Probability that exactly one flight is overbooked: {P_exactly_one_overbooked}")
```

## Exercise 5
You have three bags that each contain 100 marbles:

- Bag 1 has 75 red and 25 blue marbles;
- Bag 2 has 60 red and 40 blue marbles;
- Bag 3 has 45 red and 55 blue marbles.

You choose one of the bags at random and then pick a marble from the chosen bag, also at random. 

What is the probability that the chosen marble is red?


```python
# Your code here
# Probabilities of choosing each bag
P_bag1 = 1/3
P_bag2 = 1/3
P_bag3 = 1/3

# Probabilities of drawing a red marble from each bag
P_red_given_bag1 = 75/100
P_red_given_bag2 = 60/100
P_red_given_bag3 = 45/100

# Total probability of drawing a red marble
P_red = (P_bag1 * P_red_given_bag1) + (P_bag2 * P_red_given_bag2) + (P_bag3 * P_red_given_bag3)

# Print the result
print(f"Probability of drawing a red marble: {P_red}")
```
